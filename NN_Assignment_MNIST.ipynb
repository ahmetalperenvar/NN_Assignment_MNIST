# Question 1

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
assert x_train.shape == (60000, 28, 28)
assert x_test.shape == (10000, 28, 28)
assert y_train.shape == (60000,)
assert y_test.shape == (10000,)
```

```python
print(f"x_train set shape: {x_train.shape}    dimension: {x_train.ndim}    dtype: {x_train.dtype}")
print(f"y_train set shape: {y_train.shape}           dimension: {y_train.ndim}    dtype: {y_train.dtype}")
print(f"x_test  set shape: {x_test.shape}    dimension: {x_test.ndim}    dtype: {x_test.dtype}")
print(f"y_test  set shape: {y_test.shape}           dimension: {y_test.ndim}    dtype: {y_test.dtype}")
```

```python
plt.imshow(x_train[5],cmap='gray')
```

```python
num_classes = 10
input_shape = (28*28,1)
# convert class vectors to binary class matrices
y_train = tf.keras.utils.to_categorical(y_train, num_classes)
y_test = tf.keras.utils.to_categorical(y_test, num_classes)
```

```python
train_images = x_train.reshape((60000, 784))
train_images = train_images.astype('float32') / 255

test_images = x_test.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255
```

# Question 2

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.Input(shape=(784,)))
model.add(tf.keras.layers.Dense(16, activation='sigmoid'))
model.add(tf.keras.layers.Dense(10,activation = 'softmax'))
model.output_shape
```

```python
batch_size = 128
epochs = 32

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

history=model.fit(train_images, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
```

```python
score = model.evaluate(test_images, y_test, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```

```python
# Plot training & validation accuracy values
plt.plot(history.history['accuracy'],color='orange')
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'],color ='red')
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train'], loc='upper left')
plt.show()
```
